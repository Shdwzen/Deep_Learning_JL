```python
import tensorflow as tf
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
DataArr = load_iris()

Features = DataArr['data']
Targets = DataArr['target']

Targets = Targets.reshape(-1,1)

Encode = OneHotEncoder(sparse_output=False)
Targets = Encode.fit_transform(Targets).astype(int)
```

```python
Features.shape
```

```python
TrainX, TestX, TrainY, TestY = train_test_split(Features, Targets, test_size=0.3, random_state=42)

DecideTree = DecisionTreeClassifier()
DecideTree.fit(TrainX, TrainY)
Predictions = DecideTree.predict(TestX)
```

```python
accuracy_score(Predictions, TestY)
```

```python
Mdl1 = Sequential([
    Dense(16,input_shape=(4,),activation="relu"),
    Dense(16,activation="relu"),
    Dense(3,activation="softmax")
]
)

Mdl1.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])

Mdl1.fit(TrainX, TrainY, epochs=50, batch_size=5, validation_split=0.1, verbose=True)
```

```python
#Mdl1.save("Cool_model_1.h5")
Mdl1.evaluate(TestX, TestY)
```
