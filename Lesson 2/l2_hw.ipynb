```python

from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

```

```python

(TrainX, TrainY), (TestX, TestY) = mnist.load_data()

print("Train shape:", TrainX.shape)
print("Test shape:", TestX.shape)
```

```python
TrainX = TrainX.astype("float32") / 255.0
TestX = TestX.astype("float32") / 255.0
```

```python
TrainX = TrainX.reshape(TrainX.shape[0], 28 * 28)
TestX = TestX.reshape(TestX.shape[0], 28 * 28)
```

```python

model = Sequential()
model.add(Dense(128, activation='relu', input_shape=(784,)))
model.add(Dense(64, activation='relu'))
model.add(Dense(10, activation='softmax'))
# From a codegeeks blog
```

```python

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python

# Train model
history = model.fit(
    TrainX,
    TrainY,
    epochs=10,
    batch_size=128,
    validation_split=0.1
)
```

```python

# Evaluate model
test_loss, test_accuracy = model.evaluate(TestX, TestY)
print("Test accuracy:", test_accuracy)
```

```python

# Predict a sample digit
index = 0
prediction = model.predict(TestX[index].reshape(1, 784))
predicted_label = np.argmax(prediction)

plt.imshow(TestX[index].reshape(28, 28), cmap='gray')
plt.title(f"Predicted: {predicted_label}")
plt.axis('off')
```
